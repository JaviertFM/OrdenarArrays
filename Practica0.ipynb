```python
from time import time
import random as rn
import matplotlib.pyplot as plt
import pandas as pd
from numpy import *
```

```python
def Burbuja(a,n):
    for i in range(1,n):
        for j in range(0,n-i):
            if(a[j] > a[j+1]):
                k = a[j+1]
                a[j+1] = a[j]
                a[j] = k;

def Insercion(a,n):
    for i in range(1,n):
        v=a[i]
        j=i-1
        while j >= 0 and a[j] > v:
            a[j+1] = a[j]
            j=j-1
        a[j+1]=v

def Seleccion(a,n):
    for i in range(0,n-1):
        min=i
        for j in range(i+1,n):
            if a[min] > a[j]:
                min=j
        aux=a[min]
        a[min]=a[i]
        a[i]=aux


```

```python
nelem = 100
step = 100
i = 0
#Array de tiempos cuando le pasas un array ordenado al método burbuja
t1b = []
#Array de tiempos cuando le pasas un array inverso al método burbuja
t2b = []
#Array de tiempos cuando le pasas un array aleatorio al método burbuja
t3b = []


#Array de tiempos cuando le pasas un array ordenado al método inserción
t1i = []
#Array de tiempos cuando le pasas un array inversión al método inserción
t2i = []
#Array de tiempos cuando le pasas un array aleatorio al método inserción
t3i = []



#Array de tiempos cuando le pasas un array ordenado al método selección
t1s = []
#Array de tiempos cuando le pasas un array inverso al método selección
t2s = []
#Array de tiempos cuando le pasas un array aleatorio al método selección
t3s = []


#Habría que hacerlo hasta que el numero de elementos sea 2000
while nelem <= 2000 :
    #Creamos el array ordenado con el número de elementos
    a1 = []
    for i in range(0,nelem):
        a1.append(i)
    #Y lo copiamos para el metodo inserción y selección
    a1i = a1[:]
    a1s = a1[:]
    
    
    #Hacemos lo mismo pero con el inverso
    a2 = []
    for p in range(0, nelem):
        a2.append(a1[-1-p])
    
    
    a2i = a2[:]
    a2s = a2[:]
    
    
    #Ejecutamos los códigos con los distintos arrays:
    
    #Burbuja ordenado
    ti1b = time()
    Burbuja(a1, len(a1))
    t1b.append(time()-ti1b)
    
    
    #Inserción ordenado

    ti1i = time()
    Insercion(a1i, len(a1i))
    tf = time()
    t1i.append(tf-ti1i)
   


    
    #Selección ordenado
    ti1s = time()
    Seleccion(a1s, len(a1s))
    t1s.append(time()-ti1s)

    
    
    
    
    
    #Burbuja inverso
    ti2b = time()
    Burbuja(a2, len(a2))
    t2b.append(time()-ti2b)
    
    #Insercion inverso
    ti2i = time()
    Insercion(a2i, len(a2i))
    t2i.append(time()-ti2i)
    
    #Seleccion inverso
    ti2s = time()
    Seleccion(a2s, len(a2s))
    t2s.append(time()-ti2s)
    
    
    
    #Ahora para los aleatorios hacemos 10 vectores
    t31b = []
    t31i = []
    t31s = []
    
    for j in range(0,10):
        
        
        #Creamos el array aleatorio y guardamos el mismo en inserción y selección
        a3 = rn.sample(range(nelem),nelem)
        a3i = a3[:]
        a3s = a3[:]
        
        
        #Burbuja
        t31ib = time()
        Burbuja(a3, len(a3))
        t31b.append(time()-t31ib)
        
            
        
        
        #Insercion
        t31ii = time()
        Insercion(a3i, len(a3i))
        t31i.append(time()-t31ii)
        
        
        
        #Seleccion
        t31is = time()
        Seleccion(a3s, len(a3s))
        t31s.append(time()-t31is)
        
    #Calculamos la media de los tiempos obtenidos con el array de tamaño nelem en los 10 tiempos obtenidos
    
    #Para burbuja
    suma = 0
    mediab = 0
    for j in range(0,len(t31b)):
        suma = suma+t31b[j]
    mediab = suma/len(t31b)


    
    
    #Para inserción
    suma = 0
    mediai = 0
    for j in range(0,len(t31i)):
        suma = suma+t31i[j]
    mediai = suma/len(t31i)

    
    #Para selección
    suma = 0
    medias = 0
    for j in range(0,len(t31s)):
        suma = suma+t31s[j]
    medias = suma/len(t31s)

    #Y guardamos esa media en el array de tiempos de array generado aleatoriamente
    t3b.append(mediab)
    t3i.append(mediai)    
    t3s.append(medias)
    
    
    #Aumentamos el tamaño 100    
      
    nelem = nelem + step

    
```

```python
#Para dibujar las gráficas

#Dibujamos las gráficas para burbuja


#dibujar gráfica cuando le damos un array ordenado
count = list(range(100,2100, 100))

plt.title("Array ordenado en burbuja")
plt.plot(count,t1b)
plt.show()
#dibujar gráfica cuando le damos un array inverso
plt.title("Array inverso en burbuja")
plt.plot(count,t2b)
plt.show()

#dibujar gráfica cuando le damos un array aleatorio
plt.title("Array aleatorio en burbuja")
plt.plot(count,t3b)
plt.show()


#Y mostramos una gráfica con los tres, para ver su rango de tiempos
plt.title("Gráfica conjunta")
plt.plot(count,t1b, label ="Burbuja ordenado")
plt.plot(count,t2b, label ="Burbuja inverso")
plt.plot(count,t3b, label ="Burbuja aleatorio")
plt.legend(loc= "upper left")
plt.show()

```
